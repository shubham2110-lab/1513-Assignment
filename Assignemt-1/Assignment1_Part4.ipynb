```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define matrix 

```python
A = np.array([[3, 2],
              [2, 2]])
A
```

# Generate 200 random point on circle

```python
theta = np.linspace(0, 2 * np.pi, 200, endpoint = False)
x = np.cos(theta)
y = np.sin(theta)
unit_circle = np.vstack((x, y)) 
```

### Applying linear transformation with A to each point on unit circle 

```python
lin_transformation = A @ unit_circle
```

### Image of original circle and applying A on circle

```python
plt.figure(figsize=(8, 8))
plt.plot(unit_circle[0], unit_circle[1], label='Unit Circle', color='Green')
plt.plot(lin_transformation[0], lin_transformation[1], label='Linear Transformation Circle (Av)', color='Violet')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.title('Unit circle and Image under A')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# |Find the maximum value of ||Av|| and vector  achieve it

```python
norms = np.linalg.norm(lin_transformation, axis = 0)
max_norm_index_point = np.argmax(norms)
max_norm = norms[max_norm_index_point]
v_of_max = unit_circle[:, max_norm_index_point]

print(f"Maximum value of ||Av||: {max_norm:.4f}")
print(f"Vector v that achieves this maximum: {v_of_max}")
```

# Compute eigenvalues and eigenvectors for matrix A

```python
eigen_values, eigen_vectors = np.linalg.eig(A)
```

```python
print("\nEigenvalues of A: \n\n", eigen_values)
print("\nEigenvectors of A (each column is an eigenvector):\n\n", eigen_vectors)
```

### Compare v_max to the eigenvector with the largest eigenvalue

```python
max_eigenvalue_index = np.argmax(eigen_values)
principal_eigenvector_value = eigenvectors[:, max_eigenvalue_index]
principal_eigenvector_value /= np.linalg.norm(principal_eigenvector_value)
```

```python
print("\nPrincipal eigenvector (unit vector in direction of max eigenvalue): \n")
print(principal_eigenvector_value)
```
